# General Introduction
### API link: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#data points
x = [1, 2, 3, 4, 5, 6, 7] # days
y = [50, 51, 52, 48, 47, 49, 46] # tempurature
```

```python
plt.plot(x,y)
```

```python
#color changes to green
plt.plot(x,y, color = 'green')
```

```python
# changes the line width
plt.plot(x,y, color = 'green', linewidth = 10)
```

```python
#adding the line style
plt.plot(x,y, color = 'green', linewidth = 10, linestyle = 'dotted')
```

```python
#setting labels
plt.xlabel('Day')
plt.ylabel('Tempurature')
plt.title('Weather')
plt.plot(x,y, color = 'green', linewidth = 3, linestyle= 'dashdot')
```
